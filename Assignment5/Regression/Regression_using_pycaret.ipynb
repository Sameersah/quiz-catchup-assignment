## Load sample Data

```python
### load sample dataset from pycaret dataset module
from pycaret.datasets import get_data
data = get_data('gold')
```

## Create Transformation Pipline

```python
# import pycaret regression and init setup
from pycaret.regression import *
s = setup(data, target = 'Gold_T+22', session_id = 123)
```

## Using Object Oriented API

```python
# import RegressionExperiment and init the class
from pycaret.regression import RegressionExperiment
exp = RegressionExperiment()
```

```python
# check the type of exp
type(exp)
```

## Setup on experiment

```python
# init setup on exp
exp.setup(data, target = 'Gold_T+22', session_id = 123)
```

## Compare Models

```python
# compare baseline models
best = compare_models()
```

```python
best = create_model('knn') #choose knn to avoid plot errors with some models
```

## Analyse

```python
# plot residuals
plot_model(best, plot = 'residuals')
```
